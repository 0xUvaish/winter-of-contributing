CONVERTING THE DATASET INTO NUMPY ARRAYS AND SPLITTING IT INTO "Dependent" AND "Independent" CATAGORY  

```python
x=df.iloc[:,0:4].values
y=df.iloc[:,-1].values
```

```python
x
```

CONVERTING CATEGORICAL VALUES TO NUMERICAL VALUES TO APPLY THE DECISION TREE MODEL

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct= ColumnTransformer([("on",OneHotEncoder(drop='first'),[3])],remainder='passthrough')
x=ct.fit_transform(x)
x
```

```python
x.shape
```

```python
plt.scatter(x[:,2],y)
```

```python
plt.scatter(x[:,3],y)
```

```python
plt.scatter(x[:,4],y)
```

SPLITTING THE DATASET INTO TRAIN AND TEST

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

APPLYING THE DECISION TREE MODEL

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor(criterion='mse',random_state=0)
```

TRAINING THE DATASET

```python
dt.fit(x_train,y_train)
```

PREDICTING THE VALUE

```python
y_pred=dt.predict(x_test)
y_pred
```

```python
y_test
```

EVALUATION OF MODEL USING R2_SCORE

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
from six import StringIO
from IPython.display  import Image
from sklearn.tree import export_graphviz
import pydotplus
dot_data=StringIO()
export_graphviz(dt,out_file=dot_data,
               filled=True,rounded=True,
               special_characters=True)
graph=pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```
