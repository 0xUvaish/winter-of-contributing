# **Blurring and Smoothing images with open CV**
Smoothing and blurring is one of the most common preprocessing steps in computer vision and image processing.

We’ll be discussing the four main smoothing and blurring options that you’ll often use in your own projects:
* Simple average blurring
* Gaussian blurring
* Median filtering
* Bilateral filtering


# **Average blurring ( cv2.blur )**
Averaging of the image is done by applying a convolution operation on the image with a normalized box filter. In convolution operation, the filter or kernel slides across an image and the average of all the pixels is found under the kernel area and replaces this average with the central element of the image.

**cv2.blur(src, ksize, dst, anchor, borderType)**




```python
# importing libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt

#Read image
img = cv2.imread('opencv_logo.png')

#Apply Average Blur to the image
blur = cv2.blur(img,(10,10))

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

# **Gaussian blurring ( cv2.GaussianBlur)**
In the gaussian blur technique, the image is convolved with a gaussian filter instead of a box or normalized filter. The Gaussian blur OpenCV function has the following syntax.

**cv2.GaussianBlur( src, dst, size, sigmaX, sigmaY = 0, borderType =BORDER_DEFAULT)**

```python
# importing libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt

#Read image
img = cv2.imread('opencv_logo.png')

#Apply Gaussian Blur to the image
blur = cv2.GaussianBlur(img, (5,5),cv2.BORDER_DEFAULT)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

# **Median blurring ( cv2.medianBlur )**
The median filter technique is very similar to the averaging filtering technique shown above. The only difference is cv2.medianBlur() computes the median of all the pixels under the kernel window and the central pixel is replaced with this median value instead of the average value.

**cv2.medianBlur( src,dst,ksize )**

```python
# importing libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt

#Read image
img = cv2.imread('opencv_logo.png')

#Apply Median Blur to the image
blur = cv2.medianBlur(img,5)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

# **Bilateral blurring ( cv2.bilateralFilter )**
A bilateral filter is a non-linear, edge-preserving, and noise-reducing smoothing filter for images. It replaces the intensity of each pixel with a weighted average of intensity values from nearby pixels. This weight can be based on a Gaussian distribution. 

**cv2.bilateralFilter ( src, dst, d, sigmaColor,sigmaSpace, borderType = BORDER_DEFAULT )**

```python
# importing libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt

#Read image
img = cv2.imread('opencv_logo.png')

#Apply Bilateral Blurring to the image
blur = cv2.bilateralFilter(img,9,75,75)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

# **Image Blurring using different sized Kernels**
We can choose the size of the kernel depending on how much we want to smoothen the image. Choosing a bigger size will be averaging over a larger area. This tends to increase the smoothening effect.

```python
# Import libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt

#Read image
image = cv2.imread('opencv_logo.png')

# making filter of 3 by 3 filled with 1 divide
# by 9 for normalization
blur_filter1 = np.ones((3, 3), np.float)/(9.0)

# making filter of 5 by 5 filled with 1 divide
# by 25 for normalization
blur_filter2 = np.ones((5, 5), np.float)/(25.0)

# making filter of 7 by 7 filled with 1 divide
# by 49 for normalization
blur_filter3 = np.ones((7, 7), np.float)/(49.0)

image_blur1 = cv2.filter2D(image, -1, blur_filter1)
image_blur2 = cv2.filter2D(image, -1, blur_filter2)
image_blur3 = cv2.filter2D(image, -1, blur_filter3)

plt.imshow(image),plt.title('Original')
plt.show()
plt.imshow(image_blur1),plt.title('3x3 average kernel')
plt.show()
plt.imshow(image_blur2),plt.title('5x5 average kernel')
plt.show()
plt.imshow(image_blur3),plt.title('7x7 average kernel')
plt.show()
```
