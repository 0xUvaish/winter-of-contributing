# Title : Imbalance Data

## - Type of content : Documentation
## - Domain : Machine Learning
## - Module : Feature Engineering and Performance Metrices

Imbalanced Data is a cause of having uneven distribution of observations among different classes. When training a model on a dataset, we may achieve over 90% accuracy but when looked at data closely, it maybe posssible that 90% of the observations belonged to a single class. This creates a haphazard problem. This may lead to a problem called underfitting or overfitting.

Most of the datasets available on the internet have this problem. Hence it becomes necessary to make the observations of the dataset even with respect to the classes. If not handled imbalanced data, the model makes a clever decision to predict each of the observations given as a majority class observation to have higher accuracy. 

## Handling Imbalanced Data 

There are various ways to handle this kind of data. Some of the ways can be:
1. Collecting more data - A bigger dataset might solve the problem of uneven distribution of observation. It might increase the observations in minority classes.
2. Using different performance metrics - Accuracy doesn't actually tell the story for the data. To have the truth for the dataset, different metrics like confusion matrix, precision, recall can be used.
3. Resampling Data - This can be done by increasing copies of the observations in the minority class or decreasing the observations in majority class

### Some of the practical ways are: 

1. Under-sampling majority class : Under-sampling the majority class will resample the majority class points in the data to make them equal to the minority class.

2. Over Sampling Minority class by duplication : Resampling the minority class to make them equal to majority class

3. Over sampling Minority class by SMOTE : Generating synthetic samples in minority class to make them even with majority.

Other steps to handle imbalance data for feature engineering are :
- Handling Outliers
- Binning
- Normalisation
- Encoding
- Handling missing values
