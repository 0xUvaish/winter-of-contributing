# How it can be done?

When we are not allowed to use the openCV library then we can actually implement the Laplacian method which makes use of the second derivative instead of using the first derivative which is used for simple filters.

#What to look for?

The Laplacian function is highly sensitive to noises and thus we should try to remove the noise in the best possible way. We can do this by using zero crossing property.
We implement the Laplacian of Gaussians which combines both the Laplacian as well as Gaussian operators which will apply gaussian smoothing first and then compute the Laplacian function.

#My Approach

First I created a 3X3 box and applied custom horizontal and vertical filters and calculated the combined edge score using vertical and horizontal scores.I also scaled the edges to improve the clarity.

________________________________________________________________________________

```python
import numpy as np                                                           #importing necessary libraries
import matplotlib.pyplot as plt

vertical_filter = [[-1,-2,-1], [0,0,0], [1,2,1]]                             #define the filters
horizontal_filter = [[-1,0,1], [-2,0,2], [-1,0,1]]

img = plt.imread('/content/test.jpg')                                        #image reading

n,m,d = img.shape                                                            #extracting dimensions from image


edges_img = img.copy()                                                       #initialize the edges_img variable/

                                                                
for row in range(3, n-2):                                                    #loop over all pixels in the image
    for col in range(3, m-2):
        
        
        local_pixels = img[row-1:row+2, col-1:col+2, 0]                      #create little local 3x3 box
        
        
        vertical_transformed_pixels = vertical_filter*local_pixels           #apply the vertical filter
        
        vertical_score = vertical_transformed_pixels.sum()/4                 #remap the vertical score
        
    
        horizontal_transformed_pixels = horizontal_filter*local_pixels       #apply the horizontal filter
        
        horizontal_score = horizontal_transformed_pixels.sum()/4             #remap the horizontal score
        
        
        edge_score = (vertical_score**2 + horizontal_score**2)**.5           #combine the horizontal and vertical scores into a total edge score
        edges_img[row, col] = [edge_score]*3            

edges_img = edges_img/edges_img.max()
plt.imshow(edges_img)
```
